### OOPS Concepts

#### Inheritance

Single Inheritance

```python
class Father:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def do_math(self):
        print(self.name, " is good at math")

class Son(Father):
    def __init__(self, name, age):
        super().__init__(name, age)

    def do_physics(self):
        print(self.name, " is good at physics")

son1 = Son("karthikeya", 24)
father1 = Son("akash", 30)
son1.do_math()
son1.do_physics()
```

Hierarchical Inheritance

```python
class Father:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def do_math(self):
        print(self.name, " is good at math")

class Son(Father):
    def __init__(self, name, age):
        super().__init__(name, age)

    def do_physics(self):
        print(self.name, " is good at physics")

class Daughter(Father):
    def __init__(self, name, age):
        super().__init__(name, age)

    def do_chemistry(self):
        print(self.name, " is good at chemistry")

son1 = Son("karthikeya", 24)
daughter1 = Daughter("manjunatha", 20)
daughter1.do_math()
daughter1.do_chemistry()
```

Multilevel Inheritance

```python
class Father:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def do_math(self):
        print(self.name, " is good at math")

class Son(Father):
    def __init__(self, name, age):
        super().__init__(name, age)

    def do_physics(self):
        print(self.name, " is good at physics")

class Grandson(Son):
    def __init__(self, name, age):
        super().__init__(name, age)

    def do_computer_science(self):
        print(self.name, "is good at computer science")

son1 = Son("karthikeya", 24)
grandson = Grandson("venkatesh", 26)
grandson.do_computer_science()
grandson.do_math()
grandson.do_physics()
```

Multiple Inheritance

```python
class Father:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def do_math(self):
        print(self.name, " is good at calculus")

class Mother:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def do_law(self):
        print(self.name, "is good at law")

    def do_math(self):
        print(self.name, "is good at vector")

class Son(Father, Mother):
    def __init__(self, name, age):
        super().__init__(name, age)

    def do_physics(self):
        print(self.name, " is good at physics")

son1 = Son("krishna", 26)
son1.do_math()
```

#### Polymorphism

Method Overloading (Not supported in Python like Java)

```python
class A:
    def do_math(self, minutes, hours):
        print(f"First think and then solve the problem for {minutes} mins")
    
    def do_math(self, minutes, hours=30):
        print(f"First think for {minutes} mins and then solve for {hours} hours")

obj = A()
obj.do_math(20)
```

Method Overriding


```python
class A:
    def do_math(self, minutes):
        print(f"First think and then solve the problem for {minutes} mins")

class B(A):
    def do_math(self, minutes, hours):
        print(f"First think and then solve the problem for {minutes} mins and {hours} hours")

obj = A()
obj.do_math(20)

obj = B()
obj.do_math(20, 30)
```

#### Abstraction (with ABS: Abstract Base Class)

Contract Creation like Interfaces in Java

```python
from abc import ABC, abstractmethod

class BikeRule(ABC):
    @abstractmethod
    def apply_brake(self):
        pass
    
    # If Method Overriden then abstract method is no longer abstract, we can create object and call
    # def apply_brake(self):
    #     print("applying brake mechanism")

class Honda(BikeRule):
    # Compulsory to implement this method
    def apply_brake(self):
        print("Honda's brake mechanism")

    def start_engine(self):
        print("Engine starting mechanism")

obj = Honda()
obj.start_engine()
obj.apply_brake()
```

#### Encapsulation

```python
class Account:
    def __init__(self, name, account_number):
        self._name = name
        self.__account_number = account_number
    
    def get_name(self):
        return self._name
    
    def get_account_number(self):
        return self.__account_number

sbi_one = Account("akash", "3328")
print(sbi_one.get_name())
print(sbi_one.get_account_number())

# Hack to access private variable (protected can also be accessed by prefixing with underscore)
# print(sbi_one._Account__account_number)

# To see all methods that can be called from a class
# dir(object)
```

#### Magic Methods (Also known as dunder methods i.e double underscore methods)

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f'Name is {self.name} and age is {self.age}'

student1 = Student("akash", 24)
print(student1)
```

#### Operator Overloading

```python
class Vector:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    
    # Overloading the PLUS operator (We can overload many operators but it should have dunder method available)
    def __add__(self, other):
        return Vector(self.first + other.first, self.second + other.second)
    
    def __str__(self):
        return f'({self.first}, {self.second})'

v1 = Vector(4, 6)
v2 = Vector(6, 9)
v3 = v1 + v2
print(v3)
```

#### Iterators

```python
s = "akash is a good boy but according to whose definition of good"
it = iter(s)
while True:
    try:
        x = next(it)
        print(x, end = "")
    except:
        print("\nAll elements are printed")
        break
```

#### Generators

```python
def square(n):
    for i in range(n):
        yield(i*i)

for i in square(10):
    print(i, end = " ")
```

#### Decorators

```python
def my_decorator(func):

    def wrapper():
        print("Before function")
        func()
        print("After function")

    return wrapper

@my_decorator
def do_math():
    print("I am solving some maths problem")

do_math()
```
